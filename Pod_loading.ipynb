```python
import simpy
import random
import matplotlib.pyplot as plt
import numpy as np



train_capacity = 2;
platform_capacity = 1000;
pod_capacity = 50;



train_arrive_time = 5;
train_leave_time = 5;


offload_rate = 1; #people per second
load_rate = 1; #people per second

onboard_range = [1,train_capacity];
offload_range = [1, train_capacity];

class Station:
    def __init__ (self, env):
        self.station = simpy.Resource(env, capacity = pod_capacity)
        self.platform = simpy.Container(env, capacity = platform_capacity, init = 0)
        self.platformdata=[];
        self.traindata =[[0,0]];

class Train(object):
    def __init__(self, env, station, onboard, num_to_offload):
        self.env=env
        self.capacity = train_capacity;
        self.onboard = onboard;
        self.num_to_offload = num_to_offload;
        
        #start run process
        self.action = env.process(self.run(station))
        
    def run(self, station):
        print('T: %.1f - Train arriving with %d passengers' %  (env.now, self.onboard))
        station.traindata.append([env.now, np.nan])
        station.traindata.append([env.now, self.onboard])
        
        with station.station.request() as req:
            yield req
            station.traindata.append([env.now, self.onboard])
            yield (env.timeout(train_arrive_time))
            print('T: %.1f - Train entered with %d passengers' % (env.now,self.onboard))
            station.traindata.append([env.now, self.onboard])
            
            #off load people
            yield env.process(self.offload(station))
            station.traindata.append([env.now, self.onboard])
            
            #load people
            yield env.process(self.load(station))
            station.traindata.append([env.now, self.onboard])
            
            
            #mark the exit of the train
            station.traindata.append([env.now+train_leave_time, self.onboard])
            station.traindata.append([env.now+train_leave_time, np.nan])

            yield self.env.timeout(train_leave_time)
            print('T: %.1f - Train leaving with %d passengers' % (env.now,self.onboard))
            #station.traindata.append([env.now, self.onboard])
            
    def offload(self, station):
        print('T: %.1f - Train offloading %d people' %  (env.now, self.num_to_offload))
        self.onboard -= self.num_to_offload;        
        yield self.env.timeout(self.num_to_offload/offload_rate)
        
    def load(self, station):
        #the number of people that can be loaded is either everyone at the station
        #or all the way till the capacity is used up
        num_to_load = min(station.platform.level, self.capacity - self.onboard)
        while num_to_load < 1:
            yield(self.env.timeout(1))
        
        print('T: %.1f - Train loading %d people' %  (env.now, num_to_load))
        self.onboard += num_to_load;
        station.platform.get(num_to_load)
        yield self.env.timeout(num_to_load/load_rate)

def train_generator(env, station, onboard_range, offload_range):
    while True:
        onboard = random.randint(*onboard_range)
        offload = random.randint(1, onboard)
        train = Train(env, station1, onboard, offload)
        #wait for some time
        yield env.timeout(random.uniform(0.25,0.5))

def passenger_generator(env, station):
    while True:
        #add between 1 and 6 passengers
        station.platform.put(random.randint(1,1))
        station.platformdata.append([env.now, station.platform.level])
        #wait for some time
        yield env.timeout(random.expovariate(1/(1/4)))

def monitor(env, station):
    while True:
        print('T: %.1f - Platform: %d' % (env.now, station.platform.level))
        yield(env.timeout(0.5))
```

```python
sim_time = 15*60;
#create simulation
env = simpy.Environment()
#create station
station1 = Station(env)

env.process(train_generator(env, station1, onboard_range, offload_range))
env.process(passenger_generator(env, station1))
#env.process(monitor(env, station1))
env.run(until=sim_time)

station1.traindata.append([sim_time,station1.traindata[-1][1]])



plt.scatter(*zip(*station1.platformdata))
plt.xlabel('time (s)')
plt.ylabel('Platform Occupancy')
plt.ylim([0,platform_capacity])


plt.plot(*zip(*station1.traindata),color='r')
plt.xlabel('time (s)')
#plt.ylabel('Train occupancy')
#plt.ylim([0,train_capacity])


plt.show()


```
